```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,10,100)
plt.figure(figsize=(10,5))

plt.subplot(2,2,1)
plt.plot(x,np.sin(x))
plt.title("Sin(x)")

plt.subplot(2,2,2)
plt.plot(x,np.cos(x))
plt.title("Cos(x)")

plt.subplot(2,2,3)
plt.plot(x,np.exp(-x))
plt.title("Exp(-x)")

plt.subplot(2,2,4)
plt.plot(x,np.tan(-x))
plt.title("Tan(x)")

plt.tight_layout()
```
